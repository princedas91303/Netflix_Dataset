Remove NA values

```python
import pandas as pd

f = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/netflix_dataset.csv')

df = pd.DataFrame(f)

df[['cast','director']] = df[['cast','director']].fillna('No Data')
df['country'] = df['country'].fillna('No')
df[['date_added','rating']] = df[['date_added','rating']].fillna('5')

print(df.isna().sum())
```

Percentage of TV shows and Movies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/netflix_dataset.csv')

totalMovies =len( df[df["type"] == 'Movie'])
totalTvshow =len( df[df["type"] == 'TV Show'])
plt.pie([totalMovies, totalTvshow])
plt.show()
print((round(totalMovies/(totalMovies+totalTvshow)*100,1),round(totalTvshow/(totalMovies+totalTvshow)*100,1)))
```

Top 10 Countries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/netflix_dataset.csv')
countries=pd.crosstab(df["country"],["type"]).sort_values(by="type",ascending=True)[-10:].reset_index()
plt.figure(figsize=(12,6))
plt.barh(countries["country"], countries["type"])
for index, value in enumerate(countries["type"]):
    plt.text(value, index, str(value))
plt.title("TOP 10 COUNTRIES WITH HIGHEST NUMBER OF MOVIES AND SHOWS",fontsize=14)
plt.show()
print(("United States", 2555))
```

Most Popular in India

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/netflix_dataset.csv')
india_content = df[df['country'] == 'India']

# 1. Popularity of content type
content_counts = india_content['type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(content_counts, labels=content_counts.index, autopct='%1.1f%%')
plt.title('Content Type Popularity in India')
plt.axis('equal')
plt.show()

# 2. Find the most popular rating type: Teens, Adults or Kids
rating_counts = india_content['rating'].value_counts()
plt.figure(figsize=(8, 6))
rating_counts.plot(kind='bar')
plt.title('Rating Type Popularity in India')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

print(('Movies','TV-MA'))
```
